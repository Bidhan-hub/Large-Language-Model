```python
from openai import OpenAI
```

```python
# Initialize client
client = OpenAI()
```

```python
# create a chat completion
response = client.chat.completions.create(model="gpt-3.5-turbo",
                                messages=[{"role": "user", "content": "Listen to your"}],
                                max_tokens = 1,  # token Size
                                n=5,  # Number of Response
                                temperature=0)  # randomization

# print the chat completion
for i in range(len(response.choices)):
    print(response.choices[i].message.content)
```

```python

### Lyric Completion Assistant Code

# Initial message history (5 dictionaries)
messages = [
    {
        "role": "system",
        "content": (
            "I am Roxette lyric completion assistant. "
            "When given a line from a song, I will provide the next line in the song."
        )
    },
    {
        "role": "user",
        "content": "I know there's something in the wake of your smile"
    },
    {
        "role": "assistant",
        "content": "I get a notion from the look in your eyes, yeah"
    },
    {
        "role": "user",
        "content": "you've built a love but that love falls apart"
    },
    {
        "role": "assistant",
        "content": "Your little piece of Heaven turns too dark"
    }
]
```

```python
# Chat loop
while True:
    user_input = input("\nUser: ")

    # Exit condition
    if user_input.lower() in ["exit", "quit"]:
        print("Chat ended.")
        break

    # Append user message
    messages.append({
        "role": "user",
        "content": user_input
    })

    # API call
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages,
        temperature=0,
        max_tokens=15
    )

    # Get assistant reply
    assistant_reply = response.choices[0].message.content

    # Print response
    print("Assistant:", assistant_reply)

    # Append assistant reply back to messages
    messages.append({
        "role": "assistant",
        "content": assistant_reply
    })
```
